##### Original List Prices Across Contract Months

```python
original_monthly_prices_2024 = df_sold_2024.groupby('Month')['OriginalListPrice'].mean().sort_index().sort_values(ascending=True).reset_index()
original_monthly_prices_2024
```

```python
original_monthly_prices_2025 = df_sold_2025.groupby('Month')['OriginalListPrice'].mean().sort_index().sort_values(ascending=True).reset_index()
original_monthly_prices_2025
```

```python
original_monthly_prices_2026 = df_sold_2026.groupby('Month')['OriginalListPrice'].mean().sort_index().sort_values(ascending=True).reset_index()
original_monthly_prices_2026
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=df_sold, x="Month", y="OriginalListPrice", hue='Year', marker='o')
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.ylabel("Avg. Original List Price")
plt.show()
```
